```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

all_obs = pd.read_csv('timeseries_spec.csv')

radius_range = (1.5, 4.0)
instruments = ['NIRISS', 'NIRSPEC', 'NIRCAM']
cutoff_date = datetime.strptime('2026-01-10 00:00:00', '%Y-%m-%d %H:%M:%S')

sample = all_obs[
    (all_obs['pl_rade'] < radius_range[1]) & 
    (all_obs['pl_rade'] > radius_range[0])
]

sample_unique = sample.drop_duplicates('planet name')
sample_unique_public = sample_unique[
    [
        datetime.strptime(d, '%Y-%m-%d %H:%M:%S') < cutoff_date 
        if d is not np.nan
        else False 
        for d in sample_unique['public release date'].to_numpy()
    ]
]
names = [t for t in sample_unique_public[[i in instruments for i in sample_unique_public['instrument']]]['pl_name']]
all_transmiss = sample_unique_public[sample_unique_public['observation type'] == 'transit']

names_transmiss = np.array([t for t in all_transmiss[[i in instruments for i in all_transmiss['instrument']]]['pl_name']])
ra = np.array([t for t in all_transmiss[[i in instruments for i in all_transmiss['instrument']]]['ra']])
dec = np.array([t for t in all_transmiss[[i in instruments for i in all_transmiss['instrument']]]['dec']])

print('total planets: ', len(names_transmiss), '\n')
print('planet names: \n')
[print(name) for name in names_transmiss];

all_transmiss.to_csv('targets.csv', columns=['pl_name', 'ra', 'dec'])
```
